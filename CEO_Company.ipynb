```python
import os
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import regex as re
from bs4 import BeautifulSoup
import requests
import re
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
stop = stopwords.words('english')
```

```python
def CEOextracter(datapath):
    f = open(datapath, 'r',encoding = 'utf-8',errors = 'ignore')
    lines = f.readlines()
    article = ""
    for line in lines:
        article += line
    document = ' '.join([i for i in article.split() if i not in stop])
    article = nlp(document)
    items = [x.text for x in article.ents if x.label_ == 'PERSON']
    items3 = [x.text for x in article.ents if x.label_ == 'ORG']
    sentences = [x for x in article.sents]
    my_counter = Counter(items)
    my_counter_2 = Counter(items3)
    items2 = list(set(items))
    items4 = list(set(items3))
    candidates = pd.DataFrame( columns = ['Capital letters', 'Space','Length', 'Frequency', 'chief in sentence', 'chief in article','CEO in sentence','CEO in article'], index = items2)
    companies = pd.DataFrame(columns = ['Capital letters', 'Space','Length', 'Frequency'],index = items4)
    
    # Fill this dataframe
    label1 = 1 if 'chief executive' in article.text or 'Chief Executive' in article.text else 0
    label2 = 1 if 'ceo' in article.text or 'CEO' in article.text else 0

    for ele in items2:
        l = len(ele)
        capital_letter = 0
        space = 0
        for i in ele:
            if i.isupper():
                capital_letter += 1
            if i.isspace():
                space += 1
        candidates.loc[ele]['Capital letters'] = capital_letter
        candidates.loc[ele]['Space'] = space
        candidates.loc[ele]['Length'] = l
        candidates.loc[ele]['Frequency'] = my_counter[ele]
        related_sentences = [str(x) for x in sentences if ele in str(x)]
        candidates.loc[ele]['chief in sentence'] = 0
        candidates.loc[ele]['CEO in sentence'] = 0
        for ele2 in related_sentences:
            if 'chief executive' in ele2 or 'Chief Executive' in ele2:
                candidates.loc[ele]['chief in sentence'] = 1
                break
        for ele2 in related_sentences:
            if 'ceo' in ele2 or 'CEO' in ele2:
                candidates.loc[ele]['CEO in sentence'] = 1
                break
        candidates.loc[ele]['chief in article'] = label1
        candidates.loc[ele]['CEO in article'] = label2
        
    for ele in items4:
        l = len(ele)
        capital_letter = 0
        space = 0
        for i in ele:
            if i.isupper():
                capital_letter += 1
            if i.isspace():
                space += 1
        companies.loc[ele]['Capital letters'] = capital_letter
        companies.loc[ele]['Space'] = space
        companies.loc[ele]['Length'] = l
        companies.loc[ele]['Frequency'] = my_counter_2[ele]
    
    f.close()
    return candidates, companies
```

```python
path1 = 'C:/Users/86139/Desktop/西北大学/IEMS308/QA/origin/2013'
path2 = 'C:/Users/86139/Desktop/西北大学/IEMS308/QA/origin/2014'

files1 = os.listdir(path1)
files2 = os.listdir(path2)
C_companies = pd.DataFrame(columns = ['Capital letters', 'Space','Length', 'Frequency'],index = ['Test'])
CEO_candidates = pd.DataFrame(columns = ['Capital letters', 'Space','Length', 'Frequency', 'chief in sentence', 'chief in article','CEO in sentence','CEO in article'], index = ['Test'])
```

```python
counter = 0
for file in files1:
    datapath = path1 + "/" + file
    single_ceo, single_company = CEOextracter(datapath)
    CEO_candidates = pd.concat([CEO_candidates, single_ceo], axis = 0)
    C_companies = pd.concat([C_companies,single_company], axis = 0)
    counter += 1
    print(counter)
    
```
